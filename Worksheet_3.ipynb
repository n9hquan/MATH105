```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/marketing_campaign.csv",sep = '\t')
data
```

```python
PhD_Education = len(data[data.Education == 'PhD']) / len(data)
PhD_Education
```

```python
data.Income.mean()
```

```python
data.Income.median()
```

```python
PhD_Education_Below_mean = len(data[(data.Education == 'PhD') & (data.Income < data.Income.mean())]) / len(data)
PhD_Education_Below_mean
```

```python
PhD_Education_Below_median = len(data[(data.Education == 'PhD') & (data.Income < data.Income.median())]) / len(data)
PhD_Education_Below_median
```

```python
Edu_Mari_crosstab = pd.crosstab(data['Education'], data['Marital_Status'], margins = True)
Edu_Mari_crosstab
```

```python
Divorced_Wine = data[data['Marital_Status'] == 'Divorced'].MntWines.mean()
Single_Wine = data[data['Marital_Status'] == 'Single'].MntWines.mean()
print(Divorced_Wine, Single_Wine)
```
