##Selecting by index

Similar to Excel, you can tell Python to select some rows and some columns from a data frame. The first way to do so is to tell Python the index of the rows (colums) using the function `iloc`. For example we can select the first 3 rows and 4 colums by the following code

```python
mydata.iloc[0:3,0:4]
```

Notice that the indexes are integers starting at 0 instead of 1. Each row has an index and each columns had a label (default to be a integer starting from 0) . If you leave one of the indexes blank it means "until the end". For example `[10:,:]` means the rows from 10 to the end and all the columns. We can access any location in this table by using the `iloc` function. We need to use the dot `.` to connect the object `mydata` with the function `iloc`. In the next examples let's print out some values in the dataframe.

```python
mydata.iloc[1,2] #value in row 1 and column 2.
```

```python
mydata.iloc[1,:] #all values in row 1
```

```python
mydata.iloc[1,0:3] # values in row 1 and columns from 0 to 3
```

```python
mydata.iloc[1,3]
```

## Selecting by conditions
Look at the "top row" of `mydata`. It contains the **labels** of the columns. You can think of these labels as the names of the variables whose values are stored in the corresponding columns, and we can call for the name when we need the values. For example, if you want to access the all values of "Income", you can call its name by the dot `.` connector as follows

```python
x = mydata.Income
x.describe()
```

Or you may also try
mydata['Income']

```python
mydata[['Income','Education']]
```

`mydata.Income` is what we called a **series** of `pandas`. It basically a list of values with fixed order. You can simply think of it as a column of values in Excel. We can use the whole series or just parts of it for any calculation in Python. The next line of code will select only the Income of ones with PhD (with "Education" value is `PhD`) and put those values into a series named `PhD_Income` by adding a condition `(mydata.Education=='PhD')`.

```python
mydata.Education
```

```python
PhD_Income = mydata.Income[(mydata.Education=='PhD')]
PhD_Income
```

We can use more than one conditions to select values with operators `==`(Equals), `!=` (not Equals),`>` (Larger), `<` (Smaller), `<=` (Smaller or equals), `>=` (Larger or equals). The following code line will select all the PhD customers' incomes between 50K and 60K inclusively.

```python
mydata.Income[(mydata.Education=='PhD')&(mydata.Income>=50000)&(mydata.Income<=60000)]
```
